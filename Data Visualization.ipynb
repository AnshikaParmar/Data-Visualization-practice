```python
from collections import Counter
import matplotlib.pyplot as plt
text= "Python powerful programming language.Python data visualization and statistics  key for data analysis"
#tokenize and count word frequencies
words= text.upper().split()
word_counts= Counter(words)

#create frequency distribution
frequencies= word_counts.values()
words_list= word_counts.keys()

#statistics measures
mean_frequency= sum(frequencies)/len(frequencies)
max_frequency= max(frequencies)
min_frequency= min(frequencies)
print(f"Mean:{mean_frequency},Max:{max_frequency},Min:{min_frequency}")

#plot
plt.bar(words_list,frequencies,color='skyblue')
plt.title('Word Frequency Distribution')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=65)
plt.show
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from textblob import TextBlob
#sample text
documents=[ "Python is an amazing programming language!",
          "I find data science challenging but rewarding","Visualization tools make it easy  to understand complex data","I dislike coding errors; they are frustrating","Learning new skills is always beneficial and exciting!"]
 
#perform sentiment analysis
sentiments=[]
for doc in documents:
    blob = TextBlob(doc)
    polarity = blob.sentiment.polarity
    sentiments.append(polarity)
    
#categorize sentiments
sentiment_categories=["Positive" if score>0 else "Negative" if score< 0 else "Neutral" for score in sentiments]

#create a dataframe
import pandas as pd
df= pd.DataFrame({"Document": documents, "Sentiment Score": sentiments,"Category": sentiment_categories})

#print sentiment analysis results
print(df)

#plot sentiment distribution
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="Category",palette="plasma")
plt.title("Sentiment Analysis Distribution")
plt.xlabel("Sentiment Category")
plt.ylabel("Number of Documents")
plt.show()

#wordcloud visualization for positive sentiments
positive_text= " ".join(df[df["Category"]=="Positive"]["Document"])
wordcloud = WordCloud(width=600, height =400,colormap='magma',background_color="white").generate(positive_text)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud, interpolation="bicubic")
plt.axis("off")
plt.title("Word Cloud of positive sentiments")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Dataset
data = {
    "City": ["Mumbai", "Delhi", "Bangalore", "Hyderabad", "Chennai", "Kolkata"],
    "Latitude": [19.0760, 28.7041, 12.9716, 17.3850, 13.0827, 22.5726],
    "Longitude": [72.8777, 77.1025, 77.5946, 78.4867, 80.2707, 88.3639],
    "Population": [20400000, 19100000, 12500000, 9800000, 8900000, 14800000]  # Approximate population values
}

# Load DataFrame
df = pd.DataFrame(data)

# Create KDE plot
plt.figure(figsize=(8, 6))
sns.kdeplot(
    x=df["Longitude"], 
    y=df["Latitude"], 
    fill=True, 
    cmap="Reds",  # Color map for density visualization
    bw_adjust=0.5  # Bandwidth adjustment for smoothness
)

# Scatter plot to mark cities
plt.scatter(df["Longitude"], df["Latitude"], color="black", label="Cities", s=50)
for i, city in enumerate(df["City"]):
    plt.text(df["Longitude"][i], df["Latitude"][i], city, fontsize=10, ha="right", color="black")

# Labels & Title
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("KDE Density Plot of Indian Cities (Population)")
plt.legend()
plt.grid(True)

# Show Plot
plt.show()
```

```python
import pandas as pd
import numpy as np
data={
    "EmployeeID":[101,102,103,104,105],
    "Name":["Alice","Bob","Charlie","David","Eve"],
    "Age":[25,np.nan,30,28,np.nan],
    "Department": ["HR","IT",np.nan,"Finance","HR"],
    "Salary":[50000,60000,np.nan,58000,np.nan],
    "YearExperience":[3,np.nan,7,5,2],
}
df=pd.DataFrame(data)
print("Original DataFrame:")
print(df)

#check missing values using isnull()
print("\nmissing values in DataFrame:")
print(df.isnull())

#check non missing values using notnull()
print("\n non-missing values in DataFrame:")
print(df.notnull())

#fill missing values using fillna()
#fill missing value age values with mean
df_filled_mean=df.copy()
df_filled_mean["Age"]=df_filled_mean["Age"].fillna(df_filled_mean["Age"].mean())
df_filled_mean["Salary"]=df_filled_mean["Salary"].fillna(0)
print("\nDataFrame after replacing missing values using fillna():")
print(df_filled_mean)

#fill missing value using replace() like replacing nan with unknown
df_replaced= df.copy()
df_replaced["Department"]=df_replaced["Department"].replace(np.nan,"Unknown")
print("\nDataFrame after replacing missing values using replace():")
print(df_replaced)

#Interpolating missing values (linear interpolation)
df_interpolated=df.copy()
df_interpolated["Age"]=df_interpolated["Age"].interpolate(method="linear")
df_interpolated["YearExperience"]=df_interpolated["YearExperience"].interpolate(method="linear")
print("\nDataFrame after interpolting missing values:")
print(df_interpolated)

#drop rows with missing values using dropna()
df_dropped=df.dropna()
print("\nDataFrame after dropping rows with missing values using dropna():")
print(df_dropped)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Generate data
data_range = pd.date_range(start='2020-01-01', periods=36, freq='M')  # Monthly data for 3 years
values = np.sin(np.linspace(0, 3 * np.pi, 36)) * 10 + np.linspace(20, 50, 36) + np.random.normal(0, 2, 36)
data = pd.DataFrame({'Date': data_range, 'Value': values})
data.set_index('Date', inplace=True)

# Display first few rows of dataset
print("Sample Time Series Data:")
print(data.head())

# Plot time series
plt.figure(figsize=(10, 6))
plt.plot(data, label='Original Data')
plt.title('Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid()
plt.show()

# Rolling mean for trend analysis
data['RollingMean'] = data['Value'].rolling(window=5).mean()
plt.figure(figsize=(10, 6))
plt.plot(data['Value'], label='Original Data', color='blue')
plt.plot(data['RollingMean'], label='Rolling Mean (Window=5)', color='red', linestyle='dashed')
plt.title('Trend Analysis with Rolling Mean')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid()
plt.show()

# Decompose the time series into components
decomposition = seasonal_decompose(data['Value'], model='additive', period=12)

# Plot decomposed components
plt.figure(figsize=(12, 8))
decomposition.plot()
plt.show()

# Extract decomposed components
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Output first few values of each component
print("\nTrend Component (First 5 values):")
print(trend.dropna().head())

print("\nSeasonal Component (First 5 values):")
print(seasonal.head())

print("\nResidual Component (First 5 values):")
print(residual.dropna().head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# Generate data
data_range = pd.date_range(start='2020-01-01', periods=36, freq='M')  # Monthly data for 3 years
values = np.sin(np.linspace(0, 3 * np.pi, 36)) * 10 + np.linspace(20, 50, 36) + np.random.normal(0, 2, 36)
data = pd.DataFrame({'Date': data_range, 'Value': values})
data.set_index('Date', inplace=True)

# Display first few rows of dataset
print("Sample Time Series Data:")
print(data.head())

# Plot time series
plt.figure(figsize=(10, 6))
plt.plot(data, label='Original Data')
plt.title('Time Series Data')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid()
plt.show()

# Rolling mean for trend analysis
window_size=5
data['RollingMean']=data['Value'].rolling(window=window_size).mean()

#plot rolling mean along original data
plt.figure(figsize=(10, 6))
plt.plot(data['Value'], label='Original Data', color='blue')
plt.plot(data['RollingMean'], label='Rolling Mean (Window=5)', color='red', linestyle='dashed')
plt.title('Trend Analysis with Rolling Mean')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid()
plt.show()

#polynomial trend fitting(degree 3)
x=np.arange(len(data))
y=data['Value'].values
coeffs=np.polyfit(x,y,deg=3)
trend_poly= np.polyval(coeffs,x)

#add polynomial trend to data
data['PolynomialTrend']= trend_poly

#plot polynomial trend
plt.figure(figsize=(10,6))
plt.plot(data['Value'],label='Orginal Data',color='blue')
plt.plot(data['PolynomialTrend'],label='Polynomial Trend(Degree 3)',color='green')
plt.title('Trend Analysis using Polynomial Fitting')
plt.xlabel('Date')
plt.ylabel('Value')
plt.legend()
plt.grid()
plt.show()
 #print polynomial coefficients
print("Polynomial Coeffients(Degree 3):",coeffs)

# Decompose the time series into components
decomposition = seasonal_decompose(data['Value'], model='additive', period=12)

# Plot decomposed components
plt.figure(figsize=(12, 8))
decomposition.plot()
plt.show()

# Extract decomposed components
trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid

# Output first few values of each component
print("\nTrend Component (First 5 values):")
print(trend.dropna().head())

print("\nSeasonal Component (First 5 values):")
print(seasonal.head())

print("\nResidual Component (First 5 values):")
print(residual.dropna().head())
```
